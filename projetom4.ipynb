Aqui estão as importações das bibliotecas necessárias para a análise de dados, incluindo pandas, numpye matplotlib.

```python
#importando bibliotecas necessárias para o projeto
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Carregando os dados de maio e junho
url_maio = 'https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2023/precos-gasolina-etanol-05.csv'
url_junho = 'https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2023/precos-gasolina-etanol-06.csv'
```

```python
maio = pd.read_csv(url_maio, sep=';')
junho = pd.read_csv(url_junho, sep=';')
```

```python
maio
```

```python
junho
```

## Filtrando os dados apenas da gasolina comum

```python

gasolina_comum_maio = maio[maio['Produto'] == 'GASOLINA'].copy()
gasolina_comum_junho = junho[junho['Produto'] == 'GASOLINA'].copy()
```

## Tratando a formatação dos dados da gasolina .replace() e .astype()

```python

gasolina_comum_maio['Valor de Venda'] = gasolina_comum_maio['Valor de Venda'].str.replace(',', '.').astype(float)
gasolina_comum_junho['Valor de Venda'] = gasolina_comum_junho['Valor de Venda'].str.replace(',', '.').astype(float)



```

## Filtrando os dados apenas do etanol

```python

etanol_maio = maio[maio['Produto'] == 'ETANOL'].copy()
etanol_junho = junho[junho['Produto'] == 'ETANOL'].copy()
```

# Tratando a formatação dos dados do etanol usando .replace() e .astype()

```python

etanol_maio['Valor de Venda'] = etanol_maio['Valor de Venda'].str.replace(',', '.').astype(float)
etanol_junho['Valor de Venda'] = etanol_junho['Valor de Venda'].str.replace(',', '.').astype(float)
```

# Como se comportaram o preço dos combustíveis durante os dois meses citados? Os valores do etanol e da gasolina tiveram uma tendência de queda ou diminuição?



## crindo o gráfico com os preços da gasolina comum e do etanol nos dois meses

```python

plt.figure(figsize=(10, 6))
plt.plot(gasolina_comum_maio['Data da Coleta'], gasolina_comum_maio['Valor de Venda'], label='Gasolina Comum - Maio')
plt.plot(gasolina_comum_junho['Data da Coleta'], gasolina_comum_junho['Valor de Venda'], label='Gasolina Comum - Junho')
plt.plot(etanol_maio['Data da Coleta'], etanol_maio['Valor de Venda'], label='Etanol - Maio')
plt.plot(etanol_junho['Data da Coleta'], etanol_junho['Valor de Venda'], label='Etanol - Junho')
plt.xlabel('Data')
plt.ylabel('Preço de Venda')
plt.title('Comportamento dos Preços da Gasolina Comum e Etanol')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


## 1. **Análise Exploratória dos Preços da Gasolina e Etanol nos Meses de Maio e Junho**

Nesse gráfico, temos uma representação visual que nos mostra como os preços da gasolina comum e do etanol se comportaram nos meses de maio e junho deste ano. As linhas apresentadas têm as seguintes cores e significados:

- Linha Azul: Preço da gasolina comum em maio.
- Linha Laranja: Preço da gasolina comum em junho.
- Linha Verde: Preço do etanol em maio.
- Linha Vermelha: Preço do etanol em junho.

**Observações:**

1. Inicialmente, tanto o preço da gasolina comum quanto o do etanol em maio começam em níveis relativamente estáveis.

2. À medida que o tempo avança, há um aumento gradual nos preços de ambos os combustíveis, conforme indicado pelo movimento ascendente das linhas.

3. A linha que representa o preço da gasolina comum (azul e laranja) parece ter um aumento mais constante e menos volátil ao longo dos meses.

4. Em contraste, o preço do etanol (verde e vermelho) apresenta flutuações mais acentuadas, com oscilações visíveis nas linhas.

5. A diferença de preço entre a gasolina comum e o etanol é notável, com a gasolina comum geralmente mantendo um valor mais alto.

**Considerações:**

- A análise do gráfico nos ajuda a perceber os padrões gerais de variação de preços ao longo dos meses.

- No entanto, é importante destacar que o gráfico não fornece informações detalhadas sobre as causas específicas dessas variações.

- Fatores como disponibilidade de matéria-prima, políticas de mercado e demanda podem influenciar as oscilações observadas nos preços.

**Conclusão:**

Este gráfico nos dá uma visão visual dos movimentos de preços da gasolina comum e do etanol em maio e junho. Evidenciamos que ambos os combustíveis apresentaram aumentos, mas a gasolina comum parece seguir uma tendência mais estável, enquanto o etanol exibe variações mais marcantes.
