```python
#importando bibliotecas necessárias para o projeto
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Carregando os dados de maio e junho
url_maio = 'https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2023/precos-gasolina-etanol-05.csv'
url_junho = 'https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2023/precos-gasolina-etanol-06.csv'
```

```python
maio = pd.read_csv(url_maio, sep=';')
junho = pd.read_csv(url_junho, sep=';')
```

## Filtrando os dados apenas da gasolina comum

```python

gasolina_comum_maio = maio[maio['Produto'] == 'GASOLINA'].copy()
gasolina_comum_junho = junho[junho['Produto'] == 'GASOLINA'].copy()
```

## Tratando a formatação dos dados da gasolina .replace() e .astype()

```python

gasolina_comum_maio['Valor de Venda'] = gasolina_comum_maio['Valor de Venda'].str.replace(',', '.').astype(float)
gasolina_comum_junho['Valor de Venda'] = gasolina_comum_junho['Valor de Venda'].str.replace(',', '.').astype(float)



```

## Filtrando os dados apenas do etanol

```python

etanol_maio = maio[maio['Produto'] == 'ETANOL'].copy()
etanol_junho = junho[junho['Produto'] == 'ETANOL'].copy()
```

# Tratando a formatação dos dados do etanol usando .replace() e .astype()

```python

etanol_maio['Valor de Venda'] = etanol_maio['Valor de Venda'].str.replace(',', '.').astype(float)
etanol_junho['Valor de Venda'] = etanol_junho['Valor de Venda'].str.replace(',', '.').astype(float)
```

# Como se comportaram o preço dos combustíveis durante os dois meses citados? Os valores do etanol e da gasolina tiveram uma tendência de queda ou diminuição?



## crindo o gráfico com os preços da gasolina comum e do etanol nos dois meses

```python

plt.figure(figsize=(10, 6))
plt.plot(gasolina_comum_maio['Data da Coleta'], gasolina_comum_maio['Valor de Venda'], label='Gasolina Comum - Maio')
plt.plot(gasolina_comum_junho['Data da Coleta'], gasolina_comum_junho['Valor de Venda'], label='Gasolina Comum - Junho')
plt.plot(etanol_maio['Data da Coleta'], etanol_maio['Valor de Venda'], label='Etanol - Maio')
plt.plot(etanol_junho['Data da Coleta'], etanol_junho['Valor de Venda'], label='Etanol - Junho')
plt.xlabel('Data')
plt.ylabel('Preço de Venda')
plt.title('Comportamento dos Preços da Gasolina Comum e Etanol')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 2. Qual o preço médio da gasolina e do etanol nesses dois meses?

## Calculando o preço médio da gasolina comum, da gasolina aditivada e do etanol nos dois meses

```python
# Calculando o preço médio da gasolina comum, da gasolina aditivada e do etanol nos dois meses
avg_price_gasolina_comum_maio = gasolina_comum_maio['Valor de Venda'].mean()
avg_price_gasolina_aditivada_maio = gasolina_adtivada_maio['Valor de Venda'].mean()
avg_price_etanol_maio = etanol_maio['Valor de Venda'].mean()

avg_price_gasolina_comum_junho = gasolina_comum_junho['Valor de Venda'].mean()
avg_price_gasolina_aditivada_junho = gasolina_adtivada_junho['Valor de Venda'].mean()
avg_price_etanol_junho = etanol_junho['Valor de Venda'].mean()

# Exibindo os preços médios
print("Preço médio da gasolina comum em maio:", avg_price_gasolina_comum_maio)
print("Preço médio da gasolina aditivada em maio:", avg_price_gasolina_aditivada_maio)
print("Preço médio do etanol em maio:", avg_price_etanol_maio)

print("Preço médio da gasolina comum em junho:", avg_price_gasolina_comum_junho)
print("Preço médio da gasolina aditivada em junho:", avg_price_gasolina_aditivada_junho)
print("Preço médio do etanol em junho:", avg_price_etanol_junho)
```

- criando um gráfico para o preço médio da Gasolina comum, Gasolina aditivada e etanol

```python
# Listas de produtos e preços médios para maio e junho
produtos = ['Gasolina Comum', 'Gasolina Aditivada', 'Etanol']
precos_medios_maio = [avg_price_gasolina_comum_maio, avg_price_gasolina_aditivada_maio, avg_price_etanol_maio]
precos_medios_junho = [avg_price_gasolina_comum_junho, avg_price_gasolina_aditivada_junho, avg_price_etanol_junho]

# Largura das barras
largura = 0.35

# Posições das barras no eixo x
posicoes_maio = range(len(produtos))
posicoes_junho = [pos + largura for pos in posicoes_maio]

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(posicoes_maio, precos_medios_maio, largura, label='Maio', color='blue')
plt.bar(posicoes_junho, precos_medios_junho, largura, label='Junho', color='orange')
plt.xlabel('Produtos')
plt.ylabel('Preço Médio')
plt.title('Preço Médio de Combustíveis nos Meses de Maio e Junho')
plt.xticks([pos + largura/2 for pos in posicoes_maio], produtos)
plt.legend()
plt.tight_layout()
plt.show()
```

## Quais os 5 estados com o preço médio da gasolina e do etanol mais caros?




```python
# Calcular os preços médios da gasolina comum, gasolina aditivada e etanol para cada estado nos meses de maio e junho
avg_prices_gasolina_comum_maio = gasolina_comum_maio.groupby(['Estado - Sigla'])['Valor de Venda'].mean().reset_index()
avg_prices_gasolina_aditivada_maio = gasolina_adtivada_maio.groupby(['Estado - Sigla'])['Valor de Venda'].mean().reset_index()
avg_prices_etanol_maio = etanol_maio.groupby(['Estado - Sigla'])['Valor de Venda'].mean().reset_index()

avg_prices_gasolina_comum_junho = gasolina_comum_junho.groupby(['Estado - Sigla'])['Valor de Venda'].mean().reset_index()
avg_prices_gasolina_aditivada_junho = gasolina_adtivada_junho.groupby(['Estado - Sigla'])['Valor de Venda'].mean().reset_index()
avg_prices_etanol_junho = etanol_junho.groupby(['Estado - Sigla'])['Valor de Venda'].mean().reset_index()

# Ordenar os preços médios em ordem decrescente e pegar os 5 estados com os preços mais altos
top_states_gasolina_comum_maio = avg_prices_gasolina_comum_maio.sort_values(by='Valor de Venda', ascending=False).head(5)
top_states_gasolina_aditivada_maio = avg_prices_gasolina_aditivada_maio.sort_values(by='Valor de Venda', ascending=False).head(5)
top_states_etanol_maio = avg_prices_etanol_maio.sort_values(by='Valor de Venda', ascending=False).head(5)

top_states_gasolina_comum_junho = avg_prices_gasolina_comum_junho.sort_values(by='Valor de Venda', ascending=False).head(5)
top_states_gasolina_aditivada_junho = avg_prices_gasolina_aditivada_junho.sort_values(by='Valor de Venda', ascending=False).head(5)
top_states_etanol_junho = avg_prices_etanol_junho.sort_values(by='Valor de Venda', ascending=False).head(5)

# Impressão dos resultados para Gasolina Comum em maio
print("Top 5 estados com o preço médio mais alto da Gasolina Comum em maio:")
print(top_states_gasolina_comum_maio.to_string(index=False))

# Impressão dos resultados para Gasolina Aditivada em maio
print("\nTop 5 estados com o preço médio mais alto da Gasolina Aditivada em maio:")
print(top_states_gasolina_aditivada_maio.to_string(index=False))

# Impressão dos resultados para Etanol em maio
print("\nTop 5 estados com o preço médio mais alto do Etanol em maio:")
print(top_states_etanol_maio.to_string(index=False))

# Impressão dos resultados para Gasolina Comum em junho
print("\nTop 5 estados com o preço médio mais alto da Gasolina Comum em junho:")
print(top_states_gasolina_comum_junho.to_string(index=False))

# Impressão dos resultados para Gasolina Aditivada em junho
print("\nTop 5 estados com o preço médio mais alto da Gasolina Aditivada em junho:")
print(top_states_gasolina_aditivada_junho.to_string(index=False))

# Impressão dos resultados para Etanol em junho
print("\nTop 5 estados com o preço médio mais alto do Etanol em junho:")
print(top_states_etanol_junho.to_string(index=False))



```

```python
import matplotlib.pyplot as plt

# Dados para o gráfico de maio
states_maio = top_states_gasolina_comum_maio['Estado - Sigla']
prices_maio = top_states_gasolina_comum_maio['Valor de Venda']

# Dados para o gráfico de junho
states_junho = top_states_gasolina_comum_junho['Estado - Sigla']
prices_junho = top_states_gasolina_comum_junho['Valor de Venda']

# Criando os gráficos de barras
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  # Gráfico de maio
plt.bar(states_maio, prices_maio, color='blue')
plt.xlabel('Estados')
plt.ylabel('Preço Médio')
plt.title('Top 5 Estados com Maior Preço Médio da Gasolina Comum em Maio')
plt.xticks(rotation=45)
plt.tight_layout()

plt.subplot(1, 2, 2)  # Gráfico de junho
plt.bar(states_junho, prices_junho, color='green')
plt.xlabel('Estados')
plt.ylabel('Preço Médio')
plt.title('Top 5 Estados com Maior Preço Médio da Gasolina Comum em Junho')
plt.xticks(rotation=45)
plt.tight_layout()

# Exibindo os gráficos
plt.show()


```
