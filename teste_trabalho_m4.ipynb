```python
# Filtrando os dados da gasolina
gasolina_preco = df1.loc[df1['Produto'] == 'GASOLINA'].copy()
etanol_preco = df1.loc[df1['Produto']== 'ETANOL'].copy()

# Removendo vírgulas e convertendo a coluna "Valor de Venda" para números
gasolina_preco['Valor de Venda'] = gasolina_preco['Valor de Venda'].str.replace(',', '.', regex=True).astype(float)
etanol_preco['Valor de Venda'] = etanol_preco['Valor de Venda'].str.replace(',', '.', regex=True).astype(float)


# Calculando a média dos preços da gasolina
media_precos_gasolina = gasolina_preco['Valor de Venda'].mean()
media_precos_etanol = etanol_preco['Valor de Venda'].mean()


# Exibindo a média dos preços da gasolina
#print("Média dos preços da gasolina:", media_precos_gasolina)
print("Média dos preços da gasolina: {:.2f}".format(media_precos_gasolina))
print("Média dos preços da etanol: {:.2f}".format(media_precos_etanol))
```

```python
# Calculando estatísticas descritivas dos preços da gasolina
gasolina_prices_stats = gasolina_preco['Valor de Venda'].describe()

# Exibindo as estatísticas descritivas
print(gasolina_prices_stats)
```

```python
# Criando um histograma dos preços da gasolina
plt.hist(gasolina_preco['Valor de Venda'], bins=20, edgecolor='k')
plt.xlabel('Preço de Venda')
plt.ylabel('Frequência')
plt.title('Distribuição dos Preços da Gasolina')
plt.show()
```

```python

# Filtrando os dados da gasolina
gasolina_data = df1.loc[df1['Produto'] == 'GASOLINA'].copy()

# Convertendo a coluna "Data da Coleta" para tipo de dado datetime
gasolina_data['Data da Coleta'] = pd.to_datetime(gasolina_data['Data da Coleta'])

# Análise Temporal - Gráfico de Linha
plt.figure(figsize=(10, 6))
sns.lineplot(data=gasolina_data, x='Data da Coleta', y='Valor de Venda')
plt.title('Variação dos Preços da Gasolina ao Longo do Tempo')
plt.xlabel('Data da Coleta')
plt.ylabel('Valor de Venda')
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Comparação entre Produtos - Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=gasolina_data, x='Valor de Venda', y='Produto')
plt.title('Distribuição dos Preços da Gasolina por Produto')
plt.xlabel('Valor de Venda')
plt.ylabel('Produto')
plt.show()

# Análise de Distribuição - Histograma
plt.figure(figsize=(10, 6))
sns.histplot(data=gasolina_data, x='Valor de Venda', bins=20, kde=True)
plt.title('Distribuição dos Preços da Gasolina')
plt.xlabel('Valor de Venda')
plt.ylabel('Frequência')
plt.grid()
plt.show()

# Identificação de Outliers - Diagrama de Caixa
plt.figure(figsize=(10, 6))
sns.boxplot(data=gasolina_data, y='Valor de Venda')
plt.title('Identificação de Outliers nos Preços da Gasolina')
plt.ylabel('Valor de Venda')
plt.grid()
plt.show()
```
